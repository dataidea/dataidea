### Question 1:
What is the output of the following code?
```python
print(type([]) is list)
```

a) `True`  
b) `False`  
c) `Error`  
d) `None`

### Answer:
a) `True`

---

### Question 2:
Which of the following is not a valid Python data type?

a) `list`  
b) `tuple`  
c) `dictionary`  
d) `array`

### Answer:
d) `array`

---

### Question 3:
What does the `len()` function do?

a) Returns the number of characters in a string  
b) Returns the number of items in a list or tuple  
c) Returns the number of key-value pairs in a dictionary  
d) All of the above

### Answer:
d) All of the above

### Question 4:
How do you create a set in Python?

a) `my_set = {}`  
b) `my_set = []`  
c) `my_set = ()`  
d) `my_set = set()`

### Answer:
d) `my_set = set()`

### Question 5:
What is the output of the following code?
```python
x = "Python"
print(x[0])
```

a) `P`  
b) `y`  
c) `h`  
d) `n`

### Answer:
a) `P`

### Question 7:
What is the output of the following code?
```python
for i in range(3):
    print(i)
```

a) `1 2 3`  
b) `0 1 2`  
c) `0 1 2 3`  
d) `1 2 3 4`

### Answer:
b) `0 1 2`

---

### Question 8:
How do you create a function in Python?

a) `function my_func():`  
b) `def my_func:`  
c) `def my_func():`  
d) `func my_func():`

### Answer:
c) `def my_func():`
