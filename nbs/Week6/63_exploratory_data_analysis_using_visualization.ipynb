## Exploratory Data Analysis (EDA)

1. **Data Summary Statistics Visualization**:
   - Visualize basic summary statistics such as mean, median, mode, standard deviation, min, max, quartiles, etc. using plots like box plots, histograms, or bar plots.
   - Discuss any outliers or anomalies observed in the summary statistics and their potential impact on the analysis.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Display basic statistics
print(titanic.describe())

# Plotting
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# Box plot for Age
sns.boxplot(x='age', data=titanic, ax=axes[0, 0])
axes[0, 0].set_title('Box Plot of Age')

# Histogram for Fare
sns.histplot(x='fare', data=titanic, bins=20, kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Histogram of Fare')

# Bar plot for Sex
sns.countplot(x='sex', data=titanic, ax=axes[1, 0])
axes[1, 0].set_title('Bar Plot of Sex')

# Bar plot for Embarked
sns.countplot(x='embarked', data=titanic, ax=axes[1, 1])
axes[1, 1].set_title('Bar Plot of Embarked')

plt.tight_layout()
plt.show()
```

6. **Missing Values and Data Integrity**:
   - Visualize the presence of missing values in the dataset using heatmaps or bar plots.
   - Discuss the extent of missingness and potential strategies for handling missing data.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Check for missing values
missing_values = titanic.isnull()

# Heatmap for missing values
plt.figure(figsize=(10, 6))
sns.heatmap(missing_values, cbar=False, cmap='viridis')
plt.title('Presence of Missing Values in Titanic Dataset')
plt.show()

# Bar plot for missing values
missing_count = missing_values.sum()
plt.figure(figsize=(10, 6))
missing_count.plot(kind='bar', color='skyblue')
plt.title('Count of Missing Values in Titanic Dataset')
plt.xlabel('Variables')
plt.ylabel('Count of Missing Values')
plt.xticks(rotation=45)
plt.show()
```
